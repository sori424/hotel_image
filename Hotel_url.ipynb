```python
from selenium import webdriver
import time
from bs4 import BeautifulSoup as soup
import csv
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.common.keys import Keys
from selenium.webdriver import ActionChains
from selenium.webdriver.support.ui import Select

driver = webdriver.Chrome('chromedriver.exe')
driver.get('https://www.booking.com/')
hotel_search = '//input[@class="c-autocomplete__input sb-searchbox__input sb-destination__input"]'
hotel_list = '//li[@class="c-autocomplete__item sb-autocomplete__item sb-autocomplete__item-with_photo sb-autocomplete__item__item--elipsis sb-autocomplete__item--icon_revamp sb-autocomplete__item--hotel "]'
url_format = '//a[@class="hotel_name_link url"]'
hotel_url = {}
hotel_list2 = '//span[@class="search_hl_name"]'
button = '//button[@class="sb-searchbox__button "]'

# f = open('hotel_url.csv', 'a', encoding='utf8')
# writer = csv.writer(f)

df = pd.read_csv('hotel_url.csv', encoding='utf8')
csv_names = df['Hotel_Name'].values.tolist()  

for name in names:
    hotel_url[name] = ''
    
j = 0
for name in names:
    if name not in csv_names:
        input_hotel = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, hotel_search)))
        input_hotel.send_keys(name)
        try:
            hotel_lists = WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.XPATH, hotel_list2)))
            for i in hotel_lists:
                if i.text == name:
                    i.click()
                    button2 = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, button)))
                    button2.click()
                    break        
            hotel = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, url_format)))
            url = hotel.get_attribute('href')
            hotel_url[name] = url
            df2 = pd.DataFrame({'Hotel_Name':name, 'url':url}, index=[j])
            df.append(df2, ignore_index = True)
            driver.get('https://www.booking.com/')
            j += 1
            #   
        except:
            button2 = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, button)))
            button2.click()  
            #     
            try:
                hotel = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, url_format)))
                url = hotel.get_attribute('href')
                hotel_url[name] = url
                df2 = pd.DataFrame({'Hotel_Name':name, 'url':url}, index=[j])
                df.append(df2, ignore_index = True)
                driver.get('https://www.booking.com/')
                j += 1
            except:
                url = ''
                hotel_url[name] = ''
                df2 = pd.DataFrame({'Hotel_Name':name, 'url':url}, index=[j])
                df.append(df2, ignore_index = True)
                driver.get('https://www.booking.com/')
                j += 1

#     url = hotel.get_attribute('href')
#     hotel_url[name] = url
#     driver.get('https://www.booking.com/')
```
